```python
# import modules
import numpy as np
import astropy.units as u
from astropy.constants import G
from numpy.linalg import norm

# import plotting modules
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import rcParams

%matplotlib inline
```

```python
from galaxy.timecourse import TimeCourse
```

```python
tc = TimeCourse()
```

```python
MW = tc.read_sigmas_db('MW')
t = MW['t']
sigma_MW = MW['sigma']
M31 = tc.read_sigmas_db('M31')
sigma_M31 = M31['sigma']
M33 = tc.read_sigmas_db('M33')
sigma_M33 = M33['sigma']
```

```python
fig = plt.figure(figsize=(18,8))

# # set up the left subplot
# ax0 = plt.subplot(131)

# add the curves
plt.plot(t, sigma_MW, 'b-', lw=2, label='MW')
plt.plot(t, sigma_M31, 'r:', lw=2, label='M31')
plt.plot(t, sigma_M33, 'g.-', lw=2, label='M33')

n=50
# plt.plot(t[:n], sigma_MW[:n], 'bo', lw=2, label='MW')
# plt.plot(t[:n], sigma_M31[:n], 'ro', lw=2, label='M31')
# plt.plot(t[:n], sigma_M33[:n], 'go', lw=2, label='M33')
plt.legend();
```

```python
com = {}
for gname in ('MW','M31','M33'):
    com[gname] = tc.read_com_db(galaxy=gname)
```

```python
t = com['M33']['t']
```

```python
fig = plt.figure(figsize=(18,8))
subplots = (131, 132, 133)

# set up the left subplot
ax0 = plt.subplot(131)

# add the curves
ax0.plot(t, com['MW']['x'], 'b-', lw=2, label='x')
ax0.plot(t, com['MW']['y'], 'r:', lw=2, label='y')
ax0.plot(t, com['MW']['z'], 'g.-', lw=2, label='z')

# Add labels and subplot title
ax0.set_xlabel('Time (Gyr)', fontsize=20)
ax0.set_ylabel('Position (kpc)', fontsize=22)
ax0.legend(loc='upper right',fontsize='xx-large', shadow=True)
ax0.set_title('MW position', fontsize=24)

# set up the middle subplot
ax1 = plt.subplot(132)

# add the curves
ax1.plot(t, com['M31']['x'], 'b-', lw=2, label='x')
ax1.plot(t, com['M31']['y'], 'r:', lw=2, label='y')
ax1.plot(t, com['M31']['z'], 'g.-', lw=2, label='z')

# Add labels and subplot title
ax1.set_xlabel('Time (Gyr)', fontsize=20)
ax1.set_ylabel('Position (kpc)', fontsize=22)
ax1.legend(loc='upper right',fontsize='xx-large', shadow=True)
ax1.set_title('M31 position', fontsize=24)

# set up the right subplot
ax2 = plt.subplot(133)

# add the curves
ax2.plot(t, com['M33']['x'], 'b-', lw=2, label='x')
ax2.plot(t, com['M33']['y'], 'r:', lw=2, label='y')
ax2.plot(t, com['M33']['z'], 'g.-', lw=2, label='z')

# Add labels and subplot title
ax2.set_xlabel('Time (Gyr)', fontsize=20)
ax2.set_ylabel('Position (kpc)', fontsize=22)
ax2.legend(loc='upper right',fontsize='xx-large', shadow=True)
ax2.set_title('M33 position', fontsize=24);
```

```python
fig = plt.figure(figsize=(18,8))
subplots = (131, 132, 133)

# set up the left subplot
ax0 = plt.subplot(131)

# add the curves
ax0.plot(t, com['MW']['vx'], 'b-', lw=2, label='x')
ax0.plot(t, com['MW']['vy'], 'r:', lw=2, label='y')
ax0.plot(t, com['MW']['vz'], 'g.-', lw=2, label='z')

# Add labels and subplot title
ax0.set_xlabel('Time (Gyr)', fontsize=20)
ax0.set_ylabel('Velocity (km/s)', fontsize=22)
ax0.legend(loc='upper right',fontsize='xx-large', shadow=True)
ax0.set_title('MW velocity', fontsize=24)

# set up the middle subplot
ax1 = plt.subplot(132)

# add the curves
ax1.plot(t, com['M31']['vx'], 'b-', lw=2, label='x')
ax1.plot(t, com['M31']['vy'], 'r:', lw=2, label='y')
ax1.plot(t, com['M31']['vz'], 'g.-', lw=2, label='z')

# Add labels and subplot title
ax1.set_xlabel('Time (Gyr)', fontsize=20)
ax1.set_ylabel('Velocity (km/s)', fontsize=22)
ax1.legend(loc='upper right',fontsize='xx-large', shadow=True)
ax1.set_title('M31 velocity', fontsize=24)

# set up the right subplot
ax2 = plt.subplot(133)

# add the curves
ax2.plot(t, com['M33']['vx'], 'b-', lw=2, label='x')
ax2.plot(t, com['M33']['vy'], 'r:', lw=2, label='y')
ax2.plot(t, com['M33']['vz'], 'g.-', lw=2, label='z')

# Add labels and subplot title
ax2.set_xlabel('Time (Gyr)', fontsize=20)
ax2.set_ylabel('Velocity (km/s)', fontsize=22)
ax2.legend(loc='lower left',fontsize='xx-large', shadow=True)
ax2.set_title('M33 velocity', fontsize=24);
```

```python
angmom = {}
for gname in ('MW','M31','M33'):
    angmom[gname] = tc.read_angmom_db(galaxy=gname)
```
